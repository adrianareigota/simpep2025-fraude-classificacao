```python
#!pip install tensorflow

# Importação das bibliotecas
import pandas as pd
import pickle
import os
import sklearn.metrics as metrics
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.neural_network import MLPClassifier

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# PARÂMETROS
debugar = False                                         # Variável que indica se o código precisa ser verborrágico ou não
prettyconfusion = True                                  # Variável que indica se a matrix de confusão será bonita ou feia
calcularprecisaocomite = False                          # Variável que indica se o bloco de cálculo de precisãao do comitê deve ser executado

# Combinar colunas de texto em um único campo por linha
def imprimeCabecalho(txtCabecalho, intTamanho):
    print('')
    for x in range(intTamanho): print('︵‿', end='')
    print('')
    for x in range(int(((intTamanho * 2) - len(txtCabecalho)) / 2)): print('  ', end='')
    print(f'{txtCabecalho}')
    for x in range(intTamanho): print('︵‿', end='')
    print('')

# Imprime matrix de confusão
def imprimeMatrizConfusao(strTitulo, objPredictions, objYTest):
    
    if not prettyconfusion:
    
        # Matriz de confusão simples
        print(f"Matriz de Confusão - {strTitulo}:\n",
              pd.crosstab(objYTest, objPredictions, rownames=['Real'], 
                          colnames=['Predito'], margins=True, margins_name='Todos'))
    
    else:
    
        # Gera a matriz de confusão
        cm = confusion_matrix(objYTest, objPredictions)
        
        # Configuração do gráfico de calor
        sns.heatmap(
            cm,                    # Matriz de confusão
            annot=True,            # Mostra valores nas células
            fmt='d',              # Formato decimal inteiro
            cmap='Blues',         # Mapa de cores azul
            xticklabels=['Não Fraude', 'Fraude'],  # Rótulos eixo X
            yticklabels=['Não Fraude', 'Fraude']   # Rótulos eixo Y
        )
    
        # Adiciona rótulos e título
        plt.xlabel('Predito')      # Rótulo eixo X (o que o modelo previu)
        plt.ylabel('Verdadeiro')   # Rótulo eixo Y (o que realmente é)
        plt.title(f'Matriz de Confusão - {strTitulo}')  # Título do gráfico
        plt.show()

#Imprime a capivara de um df
def imprimeCapivara(inputDF):
    imprimeCabecalho('HEAD', 25)                        # Imprime título 'head'
    print(inputDF.head())                               # Imprime primeiras linhas
    imprimeCabecalho('INFO', 25)                        # Imprime título 'info'
    print(inputDF.info())                               # Imprime informações
    imprimeCabecalho('DIMENSÕES', 25)                   # Imprime título 'dimensões'
    print(f"Número de linhas: {inputDF.shape[0]}")      # Imprime # de linhas
    print(f"Número de colunas: {inputDF.shape[1]}")     # Imprime # de colunas
    imprimeCabecalho('VALORES NULOS', 25)               # Imprime título 'valores nulos'
    print(inputDF.isnull().sum())                       # Imprime # de valores nulos por coluna
    
    for amostraFraudes in [181, 5525, 5527, 5530]:
        imprimeCabecalho(f'REGISTRO {amostraFraudes}',25)                 # Imprime título 'registro 181'
        print(inputDF.loc[inputDF['job_id']==amostraFraudes])            # Imprime dados do registro

# Combinar colunas de texto em um único campo por linha
def combinar_textos(df_in):
    return (
        df_in['title'].fillna('') + ';' +
        df_in['location'].fillna('') + ';' +
        df_in['department'].fillna('') + ';' +
        df_in['salary_range'].fillna('') + ';' +
        df_in['company_profile'].fillna('') + ';' +
        df_in['description'].fillna('') + ';' +
        df_in['requirements'].fillna('') + ';' +
        df_in['benefits'].fillna('') + ';' +
        str(df_in['telecommuting']) + ';' +
        str(df_in['has_company_logo']) + ';' +
        str(df_in['has_questions']) + ';' +
        df_in['employment_type'].fillna('') + ';' +
        df_in['required_experience'].fillna('') + ';' +
        df_in['required_education'].fillna('') + ';' +
        df_in['industry'].fillna('') + ';' +
        df_in['function'].fillna('')
    )

#try:
if True:

    arquivo = "fake_job_postings.csv"
    if os.path.exists(arquivo):
    
        print('Arquivo localizado, iniciando a leitura do arquivo.')

        #if not (os.path.exists("modeloARV.pkl") and os.path.exists("modeloRL.pkl") and os.path.exists("modeloNB.pkl") and os.path.exists("modeloKNN.pkl") and os.path.exists("modeloSVM.pkl")):
        #if not (os.path.exists("modeloARV.pkl") and os.path.exists("modeloRL.pkl") and os.path.exists("modeloNB.pkl") and os.path.exists("modeloKNN.pkl") and os.path.exists("modeloSVM.pkl") and os.path.exists("modeloMLP.pkl")):
        if True: # Necessária execução para alimentar variáveis de treino e teste
        
            df = pd.read_csv(arquivo)                          # Leitura do arquivo e gravação em dataset
        
            print('Arquivo carregado, iniciando o processamento.')
            
            if debugar == True: imprimeCapivara(df)
            
            # Seleção dos atributos preditivos (features)
            atributos_preditivos = [
                'title', 'location', 'department', 'salary_range', 'company_profile', 'description', 'requirements', 'benefits', 'telecommuting',
                'has_company_logo', 'has_questions', 'employment_type', 'required_experience', 'required_education', 'industry', 'function'
            ]
            
            # Seleção de colunas de texto (string)
            campos_textuais = [ 
                'title', 'location', 'department', 'salary_range', 'company_profile', 'description', 'requirements', 'benefits',
                'employment_type', 'required_experience', 'required_education', 'industry', 'function'
            ]
        
            # Preencher campos nulos com a palavra 'unknown'
            if debugar == True: print("\nSubstituindo nulos por 'unknown'.")
            df[campos_textuais] = df[campos_textuais].fillna('unknown')
        
            if debugar == True: imprimeCapivara(df)
            
            # Define X (dados de entrada) e y (alvo)
            X = df[atributos_preditivos]
            y = df['fraudulent']
            
            # Divisão da base em treino (75%) e teste (25%)
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
            
            
            
            X_train_text = combinar_textos(X_train)
            X_test_text = combinar_textos(X_test)
    
        else:
            print('Treinamentos localizados, não é necessário abrir e processar o arquivo origem.')
    
        #----- Árvore de Decisão ------
        imprimeCabecalho('ARVORE DE DECISAO', 25)
        print("Iniciando processamento de árvore de decisão.")
    
        arquivo = "modeloARV.pkl"
        if os.path.exists(arquivo):
            print('Treinamento prévio de árvore de decisão localizado, carregando arquivo salvo.')
            with open(arquivo, "rb") as file:
                modeloARV = pickle.load(file)
        else:
    
            print('Treinamento prévio de árvore de decisão não localizado, processando.')
            
            # Árvore de Decisão - Pipeline com TF-IDF
            modeloARV = make_pipeline(
                TfidfVectorizer(stop_words='english', max_features=1000),
                tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
            )
            
            # Árvore de Decisão - Treinamento do modelo
            modeloARV.fit(X_train_text, y_train)
    
            # Árvore de Decisão - Salva o treinamento para não precisar treinar de novo
            with open(arquivo, "wb") as file:
                pickle.dump(modeloARV, file)
                print(f'{arquivo} salvo')
            
        # Árvore de Decisão - Predição e acurácia
        predictionsARV = modeloARV.predict(X_test_text)
        acuracia =accuracy_score(y_test, predictionsARV)
        print("\nAcurácia da Árvore de Decisão:", round(acuracia, 4))
    
        print(classification_report(y_test, predictionsARV, target_names=['Não Fraude', 'Fraude']))

        imprimeMatrizConfusao('Árvore de Decisão', predictionsARV, y_test)
                 
        
        #------ Regressão Logística ------
        imprimeCabecalho('REGRESSÃO LOGÍSTICA', 25)
        print("Iniciando processamento de regressão logística.")
    
        arquivo = "modeloRL.pkl"
        if os.path.exists(arquivo):
            print('Treinamento prévio de regressão logística localizado, carregando arquivo salvo.')
            with open(arquivo, "rb") as file:
                modeloRL = pickle.load(file)
        else:
    
            print('Treinamento prévio de regressão logística não localizado, processando.')
            
            # Regressão Logística - Criar o pipeline: TF-IDF
            modeloRL = make_pipeline(
                TfidfVectorizer(stop_words='english', max_features=1000),
                LogisticRegression(solver='lbfgs', max_iter=5000)
            )
            
            # Regressão Logística - Treinamento do modelo
            modeloRL.fit(X_train_text, y_train)
    
            # Regressão Logística - Salva o treinamento para não precisar treinar de novo
            with open(arquivo, "wb") as file:
                pickle.dump(modeloRL, file)
                print(f'{arquivo} salvo')
        
        # Regressão Logística - Predição e acurácia
        predictionsRL = modeloRL.predict(X_test_text)
        dic_RL = classification_report(y_test, predictionsRL, 
                                        target_names=['Real', 'Fraudulento'], 
                                        output_dict=True)
        result_RL = dic_RL['accuracy']
        print("\nAcurácia da Regressão Logística:", round(result_RL, 4))
    
        print(classification_report(y_test, predictionsRL, target_names=['Não Fraude', 'Fraude']))

        imprimeMatrizConfusao('Regressão Logística', predictionsRL, y_test)
    
               
        #------ Naive bayes ------
        imprimeCabecalho('NAIVE BAYES', 25)
        print("Iniciando processamento de naive bayes.")
    
        arquivo = "modeloNB.pkl"
        if os.path.exists(arquivo):
            print('Treinamento prévio de naive bayes localizado, carregando arquivo salvo.')
            with open(arquivo, "rb") as file:
                modeloNB = pickle.load(file)
    
        else:
    
            print('Treinamento prévio de naive bayes não localizado, processando.')
            
            # Naive Bayes - Criar o pipeline: TF-IDF
            modeloNB = make_pipeline(
                TfidfVectorizer(stop_words='english', max_features=1000),
                MultinomialNB()
            )
            
            # Naive Bayes - Treinar o modelo
            modeloNB.fit(X_train_text, y_train)
    
            # Naive Bayes - Salva o treinamento para não precisar treinar de novo
            with open(arquivo, "wb") as file:
                pickle.dump(modeloNB, file)
                print(f'{arquivo} salvo')
        
        # Naive Bayes - Predição e acurácia
        predictionsNB = modeloNB.predict(X_test_text)
        dic_NB = classification_report(y_test, predictionsNB,
                                        target_names=['Real', 'Fraudulento'],
                                        output_dict=True)
        result_NB = dic_NB['accuracy']
        print("\nAcurácia do Naive Bayes:", round(result_NB, 4))
    
        print(classification_report(y_test, predictionsNB, target_names=['Não Fraude', 'Fraude']))

        imprimeMatrizConfusao('Naive Bayes', predictionsNB, y_test)
                   
        #------ KNN ------
        imprimeCabecalho('KNN', 25)
        print("Iniciando processamento de KNN.")
    
        arquivo = "modeloKNN.pkl"
        if os.path.exists(arquivo):
            print('Treinamento prévio de KNN localizado, carregando arquivo salvo.')
            with open(arquivo, "rb") as file:
                modeloKNN = pickle.load(file)
    
        else:
    
            print('Treinamento prévio de KNN não localizado, processando.')
        
            # KNN - Criar o pipeline: TF-IDF + KNN (K=5 por padrão)
            modeloKNN = make_pipeline(
                TfidfVectorizer(stop_words='english', max_features=1000),
                KNeighborsClassifier(n_neighbors=5)
            )
            
            # KNN - Treinar o modelo
            modeloKNN.fit(X_train_text, y_train)
    
            # KNN - Salva o treinamento para não precisar treinar de novo
            with open(arquivo, "wb") as file:
                pickle.dump(modeloKNN, file)
                print(f'{arquivo} salvo')
        
        # KNN - Predição e acurácia
        predictionsKNN = modeloKNN.predict(X_test_text)
        dic_KNN = classification_report(y_test, predictionsKNN,
                                         target_names=['Real', 'Fraudulento'],
                                         output_dict=True)
        result_KNN = dic_KNN['accuracy']
        print("\nAcurácia do KNN:", round(result_KNN, 4))
    
        print(classification_report(y_test, predictionsKNN, target_names=['Não Fraude', 'Fraude']))
    
        imprimeMatrizConfusao('KNN', predictionsKNN, y_test)
        
        #------ SVM ------
        imprimeCabecalho('SVM', 25)
        print("Iniciando processamento de SVM.")
    
        arquivo = "modeloSVM.pkl"
        if os.path.exists(arquivo):
            print('Treinamento prévio de SVM localizado, carregando arquivo salvo.')
            with open(arquivo, "rb") as file:
                modeloSVM = pickle.load(file)
    
        else:
    
            # SVM - Criar o pipeline: TF-IDF
            modeloSVM = make_pipeline(
                TfidfVectorizer(stop_words='english', max_features=1000),
                svm.SVC()  # SVC = Support Vector Classification
            )
            
            # SVM - Treinar o modelo
            modeloSVM.fit(X_train_text, y_train)
    
            # SVM - Salva o treinamento para não precisar treinar de novo
            with open(arquivo, "wb") as file:
                pickle.dump(modeloSVM, file)
                print(f'{arquivo} salvo')
        
        # SVM - Previsão e acurácia
        predictionsSVM = modeloSVM.predict(X_test_text)
        dic_SVM = classification_report(y_test, predictionsSVM,
                                         target_names=['Real', 'Fraudulento'],
                                         output_dict=True)
        result_SVM = dic_SVM['accuracy']
        print("\nAcurácia do SVM:", round(result_SVM, 4))
    
        print(classification_report(y_test, predictionsSVM, target_names=['Não Fraude', 'Fraude']))
    
        imprimeMatrizConfusao('SVM', predictionsSVM, y_test)
    
        #------ MLP ------

        if False: # só para pular esse trecho sem precisar comentar tudo.
            imprimeCabecalho('MLP', 25)
            print("Iniciando processamento de MLP.")
        
            arquivo = "modeloMLP.pkl"
            if os.path.exists(arquivo):
                print('Treinamento prévio de MLP localizado, carregando arquivo salvo.')
                with open(arquivo, "rb") as file:
                    modeloMLP = pickle.load(file)
        
            else:
        
                # MLP - Criar o pipeline: TF-IDF
                modeloMLP = MLPClassifier(hidden_layer_sizes=(10,), max_iter=500, random_state=42)
                
                # Tokeniza sequências
                tokenizer = Tokenizer()
                tokenizer.fit_on_texts(X_train_text)
                X_train_text_token = tokenizer.texts_to_sequences(X_train_text)
                padded_sequences = pad_sequences(sequences, padding='post')
                
                # Embedding layer in a neural network
                embedding_layer = tf.keras.layers.Embedding(input_dim=100, output_dim=16)
                
                # MLP - Treinar o modelo
                modeloMLP.fit(X_train_text_token, y_train)
                
                # MLP - Salva o treinamento para não precisar treinar de novo
                with open(arquivo, "wb") as file:
                    pickle.dump(modeloMLP, file)
                    print(f'{arquivo} salvo')
            
            # MLP - Previsão e acurácia
            predictionsMLP = modeloMLP.predict(X_test_text)
            dic_MLP = classification_report(y_test, predictionsMLP,
                                             target_names=['Real', 'Fraudulento'],
                                             output_dict=True)
            result_MLP = dic_MLP['accuracy']
            print("\nAcurácia do MLP:", round(result_MLP, 4))
        
            print(classification_report(y_test, predictionsMLP, target_names=['Não Fraude', 'Fraude']))
        
            imprimeMatrizConfusao('MLP', predictionsMLP, y_test)

        #----- Cálcular métrica das predições do comitê
        # BLOCO PARA AVALIAÇÃO DA PRECISÃO DO COMITÊ   
        if calcularprecisaocomite:
           
            imprimeCabecalho('COMITÊ', 25)
            print("Iniciando processamento do COMITÊ.")
            for sensibilidade_comite in range(1, 6, 1):
                    
                predictionsCOMITE = []
            
                #print('A R N K S | T | C') 
                for vaga in X_test_text:
                    # Inicia contadores
                    tot_real = 0
                    tot_fraud = 0
                        
                    # Árvore de Decisão
                    pred_arv = modeloARV.predict([vaga])[0]
                    tot_real += int(pred_arv == 0)
                    tot_fraud += int(pred_arv == 1)
                    
                    # Regressão Logística
                    pred_rl = modeloRL.predict([vaga])[0]
                    tot_real += int(pred_rl == 0)
                    tot_fraud += int(pred_rl == 1)
                    
                    # Naive Bayes
                    pred_nb = modeloNB.predict([vaga])[0]
                    tot_real += int(pred_nb == 0)
                    tot_fraud += int(pred_nb == 1)
                    
                    # KNN
                    pred_knn = modeloKNN.predict([vaga])[0]
                    tot_real += int(pred_knn == 0)
                    tot_fraud += int(pred_knn == 1)
                    
                    # SVM
                    pred_svm = modeloSVM.predict([vaga])[0]
                    tot_real += int(pred_svm == 0)
                    tot_fraud += int(pred_svm == 1)
                  
                    if tot_fraud >= sensibilidade_comite:
                        pred_COMITE = 1
                    else:
                        pred_COMITE = 0
        
                    predictionsCOMITE.append(pred_COMITE)
                    
                    #print(f'{pred_arv} {pred_rl} {pred_nb} {pred_knn} {pred_svm} | {tot_fraud} | {pred_COMITE}') 
                
                # COMITÊ - Previsão e acurácia
                dic_COMITE = classification_report(y_test, predictionsCOMITE,
                                                 target_names=['Real', 'Fraudulento'],
                                                 output_dict=True)
                result_COMITE = dic_COMITE['accuracy']
                print(f"\nAcurácia do COMITE SENSIBILIDADE {sensibilidade_comite}:", round(result_COMITE, 4))
            
                print(classification_report(y_test, predictionsCOMITE, target_names=['Não Fraude', 'Fraude']))
            
                imprimeMatrizConfusao(f'COMITÊ SENSIBILIDADE {sensibilidade_comite}', predictionsCOMITE, y_test)

        #--------------------------------------------

        imprimeCabecalho('VALIDACAO', 25)
        
        def is_number(value):
            return value.isdigit()
        
        indice_teste = ''
        
        while indice_teste != 'x':
               
      
            # Índice da vaga a ser avaliada
            indice_teste = input("Digite o número da vaga a avaliar ou 'x' para sair: ")
            
            if is_number(indice_teste):
            
                # Exibe todos os atributos da vaga (linha completa de X_test)
        
                dfamostra = df.loc[df['job_id']==int(indice_teste)]   # Copia o registro da base desejado
                is_fraude = dfamostra['fraudulent']                   # Salva o status real da vaga (se é ou não fraude)
                dfamostra = dfamostra[atributos_preditivos]           # Deixa somente as colunas de interesse
                
                if debugar == True: print("📄 Dados da vaga selecionada antes da combinação de texto:\n")
                if debugar == True: print(dfamostra)
    
                amostra_combinada = combinar_textos(dfamostra)                # Combina os campos para ficar da forma que o modelo espera
    
                amostra = amostra_combinada.iloc[0]
    
                if debugar == True: print("📄 Dados da vaga selecionada após combinação de texto:\n")
                if debugar == True: print(amostra)
                
                # Inicia contadores
                tot_real = 0
                tot_fraud = 0
                
                # Árvore de Decisão
                pred_arv = modeloARV.predict([amostra])[0]
                tot_real += int(pred_arv == 0)
                tot_fraud += int(pred_arv == 1)
                
                # Regressão Logística
                pred_rl = modeloRL.predict([amostra])[0]
                tot_real += int(pred_rl == 0)
                tot_fraud += int(pred_rl == 1)
                
                # Naive Bayes
                pred_nb = modeloNB.predict([amostra])[0]
                tot_real += int(pred_nb == 0)
                tot_fraud += int(pred_nb == 1)
                
                # KNN
                pred_knn = modeloKNN.predict([amostra])[0]
                tot_real += int(pred_knn == 0)
                tot_fraud += int(pred_knn == 1)
                
                # SVM
                pred_svm = modeloSVM.predict([amostra])[0]
                tot_real += int(pred_svm == 0)
                tot_fraud += int(pred_svm == 1)
    
                # MLP
                #pred_mlp = modeloMLP.predict([amostra])[0]
                #tot_real += int(pred_mlp == 0)
                #tot_fraud += int(pred_mlp == 1)
                
                # Resultado da votação
                print(f"\n🗳️ Votação do comitê: Real={tot_real}, Fraudulenta={tot_fraud}\n")
                
                if tot_fraud >= 1:
                    print("🔴 O comitê acredita que a vaga é FRAUDULENTA.")
                #elif tot_fraud >= 1:
                #    print("🟡 O comitê acredita que a vaga é SUSPEITA.")
                else:
                    print("🟢 O comitê acredita que a vaga é REAL.")
                
                if tot_real == 5 or tot_fraud == 5:
                    print("✅ O comitê foi UNÂNIME na decisão.")
                else:
                    print("⚠️ O comitê NÃO foi unânime na decisão.")

                if is_fraude.item():
                    print("\nA vaga é FRAUDULENTA.")
                    if (tot_fraud >= 1): print("O comitê ACERTOU.")
                    else: print("O comitê ERROU.")
                else:
                    print("\nA vaga é REAL.")
                    if (tot_fraud >= 1): print("O comitê ERROU.")
                    else: print("O comitê ACERTOU.")            
    
    else:
        print(f"Não achei o arquivo {arquivo}")
        
#except Exception as e:
#    print(f"Erro: {e}")
```
